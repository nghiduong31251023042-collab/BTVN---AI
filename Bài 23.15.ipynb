```python
!pip install pandas folium scikit-learn -q
import pandas as pd
import folium
from sklearn.cluster import KMeans
from IPython.display import HTML, display

# ======================
# 1. Dữ liệu khách hàng
# ======================
customers = pd.DataFrame([
    {"id": 1, "lat": 10.7612, "lon": 106.6571},
    {"id": 2, "lat": 10.7629, "lon": 106.6630},
    {"id": 3, "lat": 10.7655, "lon": 106.6601},
    {"id": 4, "lat": 10.7595, "lon": 106.6549},
    {"id": 5, "lat": 10.7660, "lon": 106.6660},
    {"id": 6, "lat": 10.7582, "lon": 106.6615},
    {"id": 7, "lat": 10.7638, "lon": 106.6565},
    {"id": 8, "lat": 10.7607, "lon": 106.6645},
])

print("Dữ liệu khách hàng:")
print(customers)

# ======================
# 2. Chọn dữ liệu tọa độ
# ======================
X = customers[["lat", "lon"]]

# ======================
# 3. Dùng KMeans chia 2 cụm
# ======================
model = KMeans(n_clusters=2, random_state=42, n_init=10)
customers["cluster"] = model.fit_predict(X)
customers["cluster"] = customers["cluster"].astype(int)

# Lấy tâm cụm
centers = model.cluster_centers_

# Tạo bảng hub đề xuất
hubs = pd.DataFrame(centers, columns=["lat", "lon"])
hubs["hub_name"] = ["Hub 1", "Hub 2"]

print("\nKhách hàng sau khi chia cụm:")
print(customers)

print("\nVị trí hub đề xuất:")
print(hubs)

# ======================
# 4. Tạo bản đồ
# ======================
m = folium.Map(location=[10.762, 106.660], zoom_start=15)

colors = ["red", "blue"]

# Vẽ khách hàng
for _, row in customers.iterrows():
    folium.CircleMarker(
        location=[row["lat"], row["lon"]],
        radius=7,
        color=colors[int(row["cluster"])],
        fill=True,
        fill_color=colors[int(row["cluster"])],
        popup=f"Khách {row['id']} - Cụm {int(row['cluster'])}"
    ).add_to(m)

# Vẽ hub đề xuất
for _, row in hubs.iterrows():
    folium.Marker(
        location=[row["lat"], row["lon"]],
        popup=row["hub_name"],
        icon=folium.Icon(color="green")
    ).add_to(m)

display(HTML(m._repr_html_()))
```
