```python
!pip install osmnx networkx matplotlib -q
import osmnx as ox
import matplotlib.pyplot as plt

# Chọn khu vực
place_name = "District 1, Ho Chi Minh City, Vietnam"

# Tải mạng lưới giao thông
G = ox.graph_from_place(place_name, network_type="drive")

# Chuyển sang GeoDataFrame
nodes, edges = ox.graph_to_gdfs(G)

# In thông tin cơ bản
print("Số nút giao:", len(nodes))
print("Số cạnh đường:", len(edges))
print("Chiều dài trung bình đoạn đường (m):", edges["length"].mean())

# Vẽ bản đồ mạng lưới giao thông
fig, ax = ox.plot_graph(
    G,
    node_size=5,
    edge_linewidth=0.8,
    bgcolor="white",
    show=False,
    close=False
)

plt.show()
```
