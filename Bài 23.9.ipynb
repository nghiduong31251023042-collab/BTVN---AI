```python
!pip install pandas folium scikit-learn -q
import pandas as pd
import folium
from sklearn.cluster import KMeans
from IPython.display import HTML, display

# ======================
# Dữ liệu
# ======================
customers = pd.DataFrame([
    {"id": 1, "lat": 10.7612, "lon": 106.6571},
    {"id": 2, "lat": 10.7629, "lon": 106.6630},
    {"id": 3, "lat": 10.7655, "lon": 106.6601},
    {"id": 4, "lat": 10.7595, "lon": 106.6549},
    {"id": 5, "lat": 10.7660, "lon": 106.6660},
    {"id": 6, "lat": 10.7582, "lon": 106.6615},
    {"id": 7, "lat": 10.7638, "lon": 106.6565},
    {"id": 8, "lat": 10.7607, "lon": 106.6645},
])

# ======================
# KMeans
# ======================
X = customers[["lat", "lon"]]
model = KMeans(n_clusters=3, n_init=10)
customers["cluster"] = model.fit_predict(X)

centers = model.cluster_centers_

# ======================
# Map
# ======================
m = folium.Map(location=[10.762, 106.660], zoom_start=15)

colors = ["red", "blue", "green"]

# Khách
for _, row in customers.iterrows():
    folium.CircleMarker(
        location=[row["lat"], row["lon"]],
        radius=7,
        color=colors[int(row["cluster"])],
        fill=True,
        fill_color=colors[int(row["cluster"])],
        popup=f"Khách {row['id']} - Cụm {int(row['cluster'])}"
    ).add_to(m)

# Tâm cụm
for i, c in enumerate(centers):
    folium.Marker(
        [c[0], c[1]],
        popup=f"Tâm cụm {i}",
        icon=folium.Icon(color="black")
    ).add_to(m)

# 👉 HIỂN THỊ CHẮC CHẮN
display(HTML(m._repr_html_()))
m.save("map.html")
from google.colab import files
files.download("map.html")
```
