```python
!pip install folium pandas -q
import math
import pandas as pd
import folium
from IPython.display import HTML, display

# Hàm hiển thị map
def show_map(m):
    display(HTML(m._repr_html_()))
    return m

# Hàm tính khoảng cách
def distance(lat1, lon1, lat2, lon2):
    return math.sqrt((lat1 - lat2)**2 + (lon1 - lon2)**2)

# ======================
# 1. Dữ liệu xe
# ======================
vehicles = pd.DataFrame([
    {"id": "V1", "lat": 10.762, "lon": 106.660},
    {"id": "V2", "lat": 10.764, "lon": 106.662},
    {"id": "V3", "lat": 10.760, "lon": 106.658},
])

# ======================
# 2. Dữ liệu khách
# ======================
customers = pd.DataFrame([
    {"id": 1, "lat": 10.761, "lon": 106.659},
    {"id": 2, "lat": 10.765, "lon": 106.663},
    {"id": 3, "lat": 10.759, "lon": 106.657},
    {"id": 4, "lat": 10.763, "lon": 106.661},
])

# ======================
# 3. Gán xe gần nhất
# ======================
assignments = []

for _, c in customers.iterrows():
    best_vehicle = None
    best_dist = 999

    for _, v in vehicles.iterrows():
        d = distance(c["lat"], c["lon"], v["lat"], v["lon"])

        if d < best_dist:
            best_dist = d
            best_vehicle = v["id"]

    assignments.append({
        "customer": c["id"],
        "vehicle": best_vehicle
    })

# In kết quả
df = pd.DataFrame(assignments)
print(df)

# ======================
# 4. Vẽ bản đồ
# ======================
m = folium.Map(location=[10.762, 106.660], zoom_start=15)

# Xe (màu đỏ)
for _, v in vehicles.iterrows():
    folium.Marker(
        [v["lat"], v["lon"]],
        popup=f"Xe {v['id']}",
        icon=folium.Icon(color="red")
    ).add_to(m)

# Khách (màu xanh)
for _, c in customers.iterrows():
    folium.Marker(
        [c["lat"], c["lon"]],
        popup=f"Khách {c['id']}",
        icon=folium.Icon(color="blue")
    ).add_to(m)

# Vẽ đường nối
for _, row in df.iterrows():
    c = customers[customers["id"] == row["customer"]].iloc[0]
    v = vehicles[vehicles["id"] == row["vehicle"]].iloc[0]

    folium.PolyLine(
        [[v["lat"], v["lon"]], [c["lat"], c["lon"]]],
        color="green"
    ).add_to(m)

show_map(m)
```
