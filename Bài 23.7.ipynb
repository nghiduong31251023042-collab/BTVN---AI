```python
!pip install osmnx -q
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt

# 1. Chọn khu vực (đơn giản)
place = "District 1, Ho Chi Minh City, Vietnam"

# 2. Tải bản đồ đường
G = ox.graph_from_place(place, network_type="drive")

# 3. Chọn 2 điểm (tọa độ)
start = (10.7769, 106.7009)   # điểm A
end   = (10.7820, 106.6955)   # điểm B

# 4. Tìm node gần nhất trên bản đồ
orig = ox.distance.nearest_nodes(G, X=start[1], Y=start[0])
dest = ox.distance.nearest_nodes(G, X=end[1], Y=end[0])

# 5. Tìm đường ngắn nhất
route = nx.shortest_path(G, orig, dest, weight="length")

# 6. Vẽ đường đi
ox.plot_graph_route(G, route)
```
