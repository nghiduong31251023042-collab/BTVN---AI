```python
!pip install folium pandas -q
import folium
import pandas as pd
from folium.plugins import HeatMap
from IPython.display import HTML, display

# Dữ liệu mẫu: lat, lon, value
transaction_points = pd.DataFrame([
    {"name": "Điểm 1", "lat": 10.7630, "lon": 106.6610, "value": 120},
    {"name": "Điểm 2", "lat": 10.7610, "lon": 106.6590, "value": 200},
    {"name": "Điểm 3", "lat": 10.7645, "lon": 106.6640, "value": 80},
    {"name": "Điểm 4", "lat": 10.7598, "lon": 106.6560, "value": 300},
    {"name": "Điểm 5", "lat": 10.7652, "lon": 106.6575, "value": 150},
    {"name": "Điểm 6", "lat": 10.7602, "lon": 106.6635, "value": 240},
])

center = [10.762622, 106.660172]

m = folium.Map(location=center, zoom_start=16)

# Heatmap
heat_data = transaction_points[["lat", "lon", "value"]].values.tolist()
HeatMap(heat_data, radius=25, blur=15, max_zoom=17).add_to(m)

# Marker từng điểm
for _, row in transaction_points.iterrows():
    folium.Marker(
        [row["lat"], row["lon"]],
        popup=f"{row['name']} | value={row['value']}"
    ).add_to(m)

display(HTML(m._repr_html_()))
```
